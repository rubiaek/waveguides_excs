```python
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [20, 20]
plt.rcParams.update({'font.size': 15})
import scipy.optimize
```

# Define two fuction

```python
def fun1(x):
    return 10*np.exp(-x)
    
def fun2(x):
    return 3*np.sin(2*np.pi*x) + x
```

```python
#  Graph the functions in the same window

x = np.linspace(0,7,10000)

plt.plot(x, fun1(x), 'b', label = r'$10 e^{-x}$')
plt.plot( x, fun2(x), 'r', label = r'$3 sin(2\pi x)+x$')
plt.grid()
plt.legend(loc='upper right')
plt.title('Finding Intersections of Functions')
plt.xlabel('Input Values (x)')
plt.ylabel('Ouput Values (f)')
plt.show()
```

#  Find the points of intersection with fsolve
fsolve-Solve system of nonlinear equations Solves a problem specified by F(x) = 0 The points of intersection between the two functions are the points, which the two functions' subtraction values are zero.

```python
def diffFun(x):
    return fun1(x) - fun2(x)

plt.plot(x, diffFun(x))
plt.grid()
plt.title('Functions subtraction')
```

```python
x0 = np.linspace(1, 2.5, 5) # starts at x0
Intersection = scipy.optimize.fsolve(diffFun, x0)
print(Intersection)
```

# Draw the points of intersection between the functions


```python
plt.plot(x, fun1(x), 'b', label = r'$10 e^{-x}$')
plt.plot( x, fun2(x), 'r', label = r'$3 sin(2\pi x)+x$')
plt.plot(Intersection, fun1(Intersection), 'ko',ms = 10, label = 'Intersection with fsolve')
plt.grid()
plt.legend(loc='upper right')
plt.title('Finding Intersections of Functions')
plt.xlabel('Input Values (x)')
plt.ylabel('Ouput Values (f)')
plt.show()
```

# Find the points of intersection with out matlab function

```python
x = np.linspace(0,7,10000)

fun1x = fun1(x)
fun2x = fun2(x)
IntersectionPlace = np.where(np.diff(np.sign(fun1x-fun2x))!=0)
Intersection2 = x[IntersectionPlace]
print(Intersection2)
```
